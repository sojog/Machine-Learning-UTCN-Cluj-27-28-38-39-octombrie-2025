### Regresia simpla liniara poate fi implementata cu un modul numit scipy

# pip install scipy

```python
from scipy import stats
```

```python
rezultat = stats.linregress(suprafete, preturi)
rezultat
```

```python
slope = rezultat.slope
intercept = rezultat.intercept
```

```python
def f(x):
    return slope * x + intercept
```

```python
suprafete
```

```python
np.array(list(map(f, suprafete)))
```

```python
predictii =  slope * suprafete + intercept
predictii
```

```python
plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi)


plt.plot(suprafete, predictii)
```

```python
plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi)

plt.scatter(suprafete, predictii)
plt.plot(suprafete, predictii)
```
