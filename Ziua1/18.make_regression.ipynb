```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = make_regression(n_samples=10, n_features=1)
```

```python
plt.scatter(x, y)
plt.plot(x, y)
```

```python
x, y = make_regression(n_samples=10, n_features=1, noise=1)
plt.scatter(x, y)
# plt.plot(x, y)
```

```python
x, y = make_regression(n_samples=10, n_features=1, noise=5)
plt.scatter(x, y)
```

```python
x, y = make_regression(n_samples=10, n_features=1, noise=10)
plt.scatter(x, y)
```

```python
x, y = make_regression(n_samples=1000, n_features=1, noise=10)
plt.scatter(x, y)
```

```python
x = x.flatten()
x
```

```python
from scipy import stats

res = stats.linregress(x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, res.intercept + res.slope * x, c="red")
```
