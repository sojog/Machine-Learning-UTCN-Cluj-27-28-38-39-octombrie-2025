## Date din prima zi

```python
# m2 vs pret 
apartamente = np.array([ (37, 65_000), (45, 78_000), (60, 100_000), (75, 90_000), (86, 115_000), (78, 120_000),  (90, 150_000), (100, 170_000)])

preturi = apartamente[:, 1]
suprafete = apartamente[:, 0]
```

```python
plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi);
plt.xlabel("Suprafete")
plt.ylabel("Preturi");
```

```python
model = LinearRegression()
model
```

```python
suprafete.reshape(-1, 1)
```

```python
[ [37, 1990],  [45, 1977],  [60, 2004],  [75],  [86],  [78],  [90], [100]]
```

```python
x = suprafete.reshape(-1, 1)
y = preturi

model.fit(x, y)
model
```

```python
model.coef_, model.intercept_ # a, b
```

```python
plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi);


a, b   = model.coef_, model.intercept_ 

plt.plot(suprafete, a * suprafete + b, c="red");

plt.xlabel("Suprafete")
plt.ylabel("Preturi");
```
