```python
score_dict = {}

for k in range(1, 11):
    knn_model = KNeighborsRegressor(n_neighbors=k)
    knn_model.fit(x_train, y_train)
    score = knn_model.score(x_test, y_test)
    print("K=", k, "score=", score)
    score_dict[k] = score

plt.bar(score_dict.keys(), score_dict.values());
```
