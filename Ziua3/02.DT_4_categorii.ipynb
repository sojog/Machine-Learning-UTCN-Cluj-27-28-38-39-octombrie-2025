```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/4_categories.csv")
df
```

```python
df["category"].unique()
```

```python
# x 
data = df[["coord_x", "coord_y"]]
# y 
target = df[["category"]]
```

```python
plt.scatter(data["coord_x"], data["coord_y"])
```

```python
plt.scatter(data["coord_x"], data["coord_y"], c=target["category"].map({"a":"red", "b":"blue", "c":"green", "d":"black"}))
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2)
```

```python
plt.scatter(x_train["coord_x"], x_train["coord_y"])
plt.scatter(x_test["coord_x"], x_test["coord_y"])
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(x_train, y_train)
```

```python
from sklearn.tree import export_graphviz
import graphviz
```

```python
exported_data = export_graphviz(tree_model, feature_names=["coord_x", "coord_y"], class_names=["a", "b", "c", "d"], filled=True, rounded=True)
exported_data
```

```python
tree_source = graphviz.Source(exported_data)
tree_source.render("output/4Categorii", format="png")
tree_source.view()
```
