```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/4_categories.csv")
df
```

```python
data = df[["coord_x", "coord_y"]]
target = df[["category"]]
```

```python
plt.scatter(data["coord_x"], data["coord_y"]);
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=4)
model
```

```python
model.fit(data)
```

```python
model_colors = pd.DataFrame(model.labels_, columns=["cluster"])
model_colors["cluster"] = model_colors["cluster"].map({2:"red", 0:"blue", 1:"green", 3:"black", 4:"yellow", 5:"purple"})
```

```python
plt.scatter(data["coord_x"], data["coord_y"], c=model_colors["cluster"] );
```

```python
plt.scatter(data["coord_x"], data["coord_y"], c=target["category"].map({"a":"red", "b":"blue", "c":"green", "d":"black"}))
plt.show()
```
