```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits, load_iris
digits = load_digits()
iris = load_iris()
```

```python
print(iris["DESCR"])
```

```python
iris_data = iris.data
iris_target = iris.target
```

```python
iris_data[0]
```

```python
sepal_data = iris_data[:,0:2]
```

```python
plt.scatter(sepal_data[:, 0], sepal_data[:,1], c=iris_target)
```

```python
petal_data = iris_data[:,2:4]
plt.scatter(petal_data[:, 0], petal_data[:,1], c=iris_target)
```

```python
from sklearn.cluster import KMeans
means_model = KMeans(3)
means_model.fit(petal_data)
```

```python
plt.scatter(petal_data[:, 0], petal_data[:,1], c=means_model.labels_)
```
