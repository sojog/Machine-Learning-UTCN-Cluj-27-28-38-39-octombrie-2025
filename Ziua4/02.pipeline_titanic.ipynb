## Label Encoding -> per coloanele Sex, Embarked

```python
def label_encode_column(df, column):
    label_encoder = LabelEncoder()
    df[column] = label_encoder.fit_transform(df[column])
    return df
```

```python
df_test.head(2)
```

```python
# Varianta 1 
label_encode_column(df_test, "Embarked").head(2)
```

```python
df_test = df_test.pipe(label_encode_column, "Embarked").pipe(label_encode_column, "Sex")
df_test.head(2)
```

```python
def drop_column(df, column):
    return df.drop(columns=[column])
```

```python
df_test = df_test.pipe(drop_column, "Name").pipe(drop_column, "Ticket").pipe(drop_column, "Cabin")
df_test.head(2)
```
